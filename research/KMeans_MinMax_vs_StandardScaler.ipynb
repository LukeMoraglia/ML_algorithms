```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
import sklearn.metrics as metrics
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn import datasets
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns

%matplotlib inline
#plt.rcParams['figure.figsize'] = [16, 9]
plt.style.use("fivethirtyeight")
%load_ext autoreload
%autoreload 2
```

```python
X, y = datasets.make_classification(random_state=0, 
                                    n_samples=1000,
                                    n_features=2,
                                    n_informative=2,
                                    n_redundant=0,
                                    n_classes=3,
                                    n_clusters_per_class=1,
                                    flip_y = 0,
                                    class_sep=3,
                                    scale=[1, 20])
```

```python
n_clus = 3
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X[:,0], X[:,1], c=y)
plt.title("Original data")
plt.show()
```

```python
unscaled_KM = KMeans(n_clusters=n_clus)
unscaled_KM.fit(X)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X[:,0], X[:,1], c=unscaled_KM.labels_)
plt.title(f"Original Data - Kmeans {n_clus} clusters")
plt.show()
```

```python
X_stand = StandardScaler().fit_transform(X)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X_stand[:,0], X_stand[:,1], c=y)
plt.title("StandardScaler Data")
plt.show()
```

```python
stand_KM = KMeans(n_clusters=n_clus)
stand_KM.fit(X_stand)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X_stand[:,0], X_stand[:,1], c=stand_KM.labels_)
plt.title(f"StandardScaler Data - Kmeans {n_clus} clusters")
plt.show()
```

```python
X_minmax = MinMaxScaler().fit_transform(X)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X_minmax[:,0], X_minmax[:,1], c=y)
plt.title("MinMax Data")
plt.show()
```

```python
minmax_KM = KMeans(n_clusters=n_clus)
minmax_KM.fit(X_minmax)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X_minmax[:,0], X_minmax[:,1], c=minmax_KM.labels_)
plt.title(f"Minmax Data - Kmeans {n_clus} clusters")
plt.show()
```
